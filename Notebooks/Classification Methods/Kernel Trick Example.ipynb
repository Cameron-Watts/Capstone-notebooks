```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define data for each class in one dimension
class_1 = np.array([-10, -8, -6, 6, 8, 10])
class_2 = np.array([-4, -2, 0, 2, 4])

#Dummy variables to plot on a single line
straight_line_1 = np.array([0,0,0,0,0,0])
straight_line_2 = np.array([0,0,0,0,0])

#Kernel function
def kernel(data):
    return -data**2 + 10

#Apply kernel function
kernel_data_1 = kernel(class_1)
kernel_data_2 = kernel(class_2)
```

```python
#Plot points in their single dimension
plt.scatter(class_1, straight_line_1, label="Class 1")
plt.scatter(class_2, straight_line_2, label="Class 2")

plt.legend()

plt.title("Data Points in 1-Dimension")
```

```python
#Plot kernel function
plt.plot(list(range(-10, 11)), kernel(np.array(range(-10, 11))),
         color="y", label="Kernel Function", ls="--", alpha=0.75)

#Plot points with their associated kernel function transformed y-values
plt.scatter(class_1, kernel_data_1, label="Class 1")
plt.scatter(class_2, kernel_data_2, label="Class 2")

#Plot decision boundary
plt.axhline(-16, color="r" , label="Class Prediction Boundary")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
