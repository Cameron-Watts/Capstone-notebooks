```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.power import TTestIndPower

#The range of sample sizes that the power will be determined for
sample_sizes = np.array(range(2,500))

#The effect sizes to test
effect_sizes = np.array([0.2, 0.5, 0.8])

#Alpha value (significance level or type 1 error rate)
alpha = 0.05

#Create object to use
power_calculator = TTestIndPower()

#Plot the graph of the power level against sample size
power_calculator.plot_power(dep_var = "nobs", nobs = sample_sizes,
                            effect_size=effect_sizes, alpha=alpha)

#Show the graph
plt.show()
```
